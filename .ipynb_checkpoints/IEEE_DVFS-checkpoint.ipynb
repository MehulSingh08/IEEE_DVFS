```python
import json
import numpy as np
import pandas as pd
import cvxpy as cp
import matplotlib.pyplot as plt

def energy_optimizer_from_csv():
    # Load dynamic dataset from CSV
    df = pd.read_csv('dataset.csv')

    # Extract necessary columns (mapping your dataset columns)
    cpu_power = df['allDataMean4 (CPU Power Consumption)'].values.reshape(-1, 1)
    cpu_freq = df['allDataMean5 (CPU Freq)'].values.reshape(-1, 1)
    t_ratio = df['allDataMean7 (T_sup/T_cpu)'].values.reshape(-1, 1)

    # Load static config JSON
    with open('dc_energy_input.json', 'r') as file:
        dc_energy_input = json.load(file)

    m = dc_energy_input["number_of_machines"]
    t_red = dc_energy_input["red_temperature"]
    c = dc_energy_input["heat_capacity"]
    p = dc_energy_input["air_density"]
    s = dc_energy_input["flow_speed"]
    A = np.array(dc_energy_input["airflow_rec_correlation"])

    cpu_max_freq = dc_energy_input["machine_max_frequency"]
    cpu_max_power = dc_energy_input["cpu_max_power"]
    cpu_idle_power = dc_energy_input["cpu_idle_power"]
    cpu_mass = dc_energy_input["cpu_mass"]
    cpu_heat_capacity = dc_energy_input["cpu_heat_capacity"]

    mass_of_air_in = p * s  # g/s
    rec_coefficient = np.transpose(np.sum(A, axis=0, keepdims=True))
    AT = np.transpose(A)
    rev_EAT = np.linalg.inv(np.eye(m) - AT)
    m_sup_array = np.multiply(mass_of_air_in, (1 - rec_coefficient))
    alpha = c * m_sup_array

    c_0 = cpu_idle_power ** 0.5
    c_1 = (cpu_max_power ** 0.5 - c_0) ** 0.5 / cpu_max_freq
    P_red = np.multiply(t_red * c * mass_of_air_in, np.ones((m, 1)))
    X_max = cpu_max_power ** 0.5

    # Optimization Variables
    x = cp.Variable((m, 1))
    P_e = cp.Variable((m, 1))
    COP = cp.Variable((1, 1))
    P_sup = cp.Variable((m, 1))
    T_sup = cp.Variable((1, 1))
    f = cp.Variable((m, 1))

    objective = cp.Minimize(cp.quad_over_lin(x, COP) + cp.sum(P_e))

    constraints = [
        x >= cp.power(c_1 * f, 2) + c_0,
        x <= X_max,
        P_e >= cp.power(x, 2),
        P_e <= cpu_max_power,
        T_sup * alpha == P_sup,
        T_sup >= 5.5,
        COP == 0.265 * T_sup - 1.45,
        P_sup >= 0,
        rev_EAT @ (P_sup + P_e) <= P_red,
        f <= cpu_max_freq
    ]

    # Use CPU Frequency from dataset as load baseline
    constraints.append(f == cpu_freq)

    prob = cp.Problem(objective, constraints)
    result = prob.solve()

    # Post-calculation
    p_out = rev_EAT @ (P_sup.value + P_e.value)
    T_in = (P_sup.value + AT @ p_out) / (c * (m_sup_array + AT @ (p * s * np.ones((m, 1)))))
    T_cpu = T_in + P_e.value / (cpu_heat_capacity * cpu_mass)
    T_out = p_out / (c * (p * s))
    P_AC = sum(P_e.value) / COP.value
    P_CMP = sum(P_e.value)
    P_TOTAL = P_AC + P_CMP

    # Print Results
    print(f"AC Supply Temperature (Optimized): {T_sup.value[0][0]:.2f}°C")
    print(f"Total AC Power: {P_AC[0][0]:.2f} W")
    print(f"Total Compute Power: {P_CMP[0][0]:.2f} W")
    print(f"Total Energy Consumption: {P_TOTAL[0][0]:.2f} W")

    # Plot graphs
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.plot(cpu_power, label='CPU Power Consumption')
    plt.xlabel('Sample')
    plt.ylabel('Power (W)')
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(cpu_freq, color='orange', label='CPU Frequency')
    plt.xlabel('Sample')
    plt.ylabel('Frequency (MHz)')
    plt.legend()

    plt.subplot(2, 2, 3)
    plt.plot(T_cpu, color='red', label='CPU Temperature')
    plt.xlabel('Machine')
    plt.ylabel('Temperature (°C)')
    plt.legend()

    plt.subplot(2, 2, 4)
    plt.plot(t_ratio, color='green', label='T_sup/T_cpu Ratio')
    plt.xlabel('Sample')
    plt.ylabel('Ratio')
    plt.legend()

    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    energy_optimizer_from_csv()
```
